```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import roc_auc_score,roc_curve,classification_report,confusion_matrix,plot_confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=111)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
y_train.value_counts(1)
```

```python
y_test.value_counts(1)
```

```python
model_LR = LogisticRegression(solver = 'saga',max_iter = 10000,penalty='none',verbose=True,n_jobs=2)
```

```python
model_LR.fit(X_train,y_train)
```

```python
ytrain_predict = model_LR.predict(X_train)
ytest_predict = model_LR.predict(X_test)
```

```python
ytest_predict_prob = model_LR.predict_proba(X_test)
pd.DataFrame(ytest_predict_prob).head()
```

```python
model_LR.score(X_train,y_train)
```

```python
probs = model_LR.predict_proba(X_train)
probs = probs[:,1]
auc = roc_auc_score(y_train,probs)
auc
```

```python
train_fpr,train_tpr,train_thresholds = roc_curve(y_train,probs)
plt.plot([0,1],[0,1],linestyle='--')
plt.plot(train_fpr,train_tpr)
```

```python
probs = model_LR.predict_proba(X_train)
probs = probs[:,1]
auc = roc_auc_score(y_train,probs)
print(auc)

train_fpr,train_tpr,train_thresholds = roc_curve(y_train,probs)
plt.plot([0,1],[0,1],linestyle='--');
plt.plot(train_fpr,train_tpr);
```

```python
model_LR.score(X_test,y_test)
```

```python
probs = model_LR.predict_proba(X_test)
probs = probs[:,1]
auc = roc_auc_score(y_test,probs)
print(auc)

test_fpr,test_tpr,test_thresholds = roc_curve(y_test,probs)
plt.plot([0,1],[0,1],linestyle='--');
plt.plot(test_fpr,test_tpr);
```

```python
grid = {'penalty':['l2','none'],
          'solver':['saga','lbfgs'], #‘newton-cg’, ‘lbfgs’, ‘liblinear’, ‘sag’, ‘saga’
           'tol':[0.0001,0.00001]}
```
